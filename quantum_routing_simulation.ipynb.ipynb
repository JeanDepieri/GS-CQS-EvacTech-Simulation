```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer 
from qiskit.visualization import plot_histogram
from matplotlib import pyplot as plt

circuito = QuantumCircuit(2, 2)  # qubits (rotas), 2 bits clássicos (medidas)

# Aplicar porta Hadamard ao qubit 0 para criar superposição
circuito.h(0)

# Aplicar porta CNOT para emaranhar os qubits (rota 0 influencia rota 1)
circuito.cx(0, 1)

# Medição dos qubits nos registradores clássicos
circuito.measure([0,1], [0,1])

# Simular o circuito quântico
circuito.draw('mpl')
```
